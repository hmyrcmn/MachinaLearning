```python
y=data.iloc[:,13:].values#y bagımlı değişken .
y
```

```python
from sklearn import preprocessing

le =preprocessing.LabelEncoder()

x[:,1]=le.fit_transform(x[:,1]) #kategorik değişken olan georapy sayısal bir değere dönüştürüldü
x
```

```python
le2 =preprocessing.LabelEncoder()

x[:,2]=le2.fit_transform(x[:,2]) #kategorik değişken olan cinsiyet  sayısal bir değere dönüştürüldü kadın 0 erkek 1 olarak düzenlendi
x
```

```python
# one hot encoding ile 0 1 2 degerlerine dönüştürülen kategorik değişken 0 -1 ile temsil edilir .
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

ohe=ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(sparse=False, dtype=float), [1])],
    remainder='passthrough'
)

x=ohe.fit_transform(x)
x[:,1]

x
```

kategorik verileri  ülke tr fr us ın işlenebilecek formata dönüştüürlülmesi

```python
#eğitim ve test verilerinin oluştrulması
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)

x_train
```

ÖZNİTEKİL ÖLÇEKLEME

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

X_train=sc.fit_transform(x_train)
X_test=sc.fit_transform(x_test)

```

yapay sinir ağı oluşturma

```python
import keras

#yağay sinir ağı oluştur komutu veriliyor keras a
from keras.models import Sequential

#katman oluştruma
from keras.layers import Dense

classifier=Sequential() # yapay sinir ağı oluşturuldu içi boş
# 6 nöron ve 11 nöron girişinden oluşan katman eklendi
classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu', input_dim=12))
  # kaç tane gizli katman veirleceği :32   11+1/2 = 6  ,nput +output

# 2 gizli katman
classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))


# çıkış katmanı
classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))


#optimizasyon :
#binary kullanılacak bagımsız değişken 0 1 den oluşuyor
classifier.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

#adam sdg :veri kümesini küçük parçalara (minibatch) bölmek ve her minibatch için gradyan hesaplamalarını yapmaktır.

```

```python
# noron network çalıştırma
classifier.fit(X_train, y_train, epochs=50)
# kaç aşamada tekrar edileceğini eposchi belirler

y_pred=classifier.predict(X_test)
y_pred
```

```python
y_pred=(y_pred>0.5) # t /f 1/0 0.5 den büyükse true
y_pred
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,y_pred)

cm
```

```python
import pickle

dosya = "DerinOgrenmeModel.kayit"

# Modeli kaydetme
with open(dosya, "wb") as file:
    pickle.dump(classifier, file)

# Modeli yükleme
with open(dosya, "rb") as file:
    loadedModel = pickle.load(file)

# Tahminlerde bulunma
predictions = loadedModel.predict(X_test)
predictions
```
