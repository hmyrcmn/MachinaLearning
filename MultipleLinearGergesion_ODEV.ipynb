```python
df = pd.DataFrame(data)
le = preprocessing.LabelEncoder()
df['windy'] = le.fit_transform(df['windy'])
df['play']=le.fit_transform(data.iloc[:,-1])

df

# data tidy true=1 yes=1 and false=0 no=0 in play and windy column in data
```

```python
#one hot encoding işlemi
df_OneEncoded = pd.get_dummies(df, columns=['outlook'])
data=df_OneEncoded
data
```

```python

# One-Hot Encoding sonrası sütun adlarını al
sutunlar = list(df_OneEncoded.columns)

# Belirtilen sıralama ile sütunları yeniden düzenle
sirali_sutunlar = ['windy', 'play'] + sutunlar[3:] + sutunlar[0:3]
df_OneEncoded = df_OneEncoded[sirali_sutunlar]

# Sonuç veriyi görüntüleme
data = df_OneEncoded
data
```
