```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer= SimpleImputer(missing_values=np.nan,strategy='mean')
Yas=data.iloc[:,1:4].values
print(Yas)

imputer =imputer.fit(Yas[:,1:4])
Yas[:,1:4]=imputer.transform(Yas[:,1:4])
print(Yas)
```

kategorik verileri  ülke tr fr us ın işlenebilecek formata dönüştüürlülmesi

```python
country=data.iloc[:,0:1].values

print(country)

# caunty tr =1, us =2 fr=0 olrak düzenlendi
from sklearn import preprocessing
# LABEL ENCODER  SSAYISAL DEĞER VEİRRİ 0 1-2-3
le=preprocessing.LabelEncoder()
country[:,0]=le.fit_transform(data.iloc[:,0])
print(country)
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
#ohe = OneHotEncoder #use import with special library

# remember OneHotEncoder uding without ()
# ONE HOT ENCODER COLON BAŞLIKLARI İLE ATAMA YAPARAK COLONA AİDİYETİN 0 -1 İLE İFADE EDİLMESİNİ SAĞLAR :
# TR COLONUND VERİ 1 DEGERİ TAŞIYORSA O KİŞİ TR  O İSE TR DEĞİL FR YA DA US
ohe=preprocessing.OneHotEncoder

country = ohe.fit_transform(country).toarray()
print("One-Hot Encoded Ülke Dizisi:")
print(country)
```
