BOY KLO YADAN CNSYET TAHMN


```python
import pandas as pd

veriler = pd.DataFrame(veriler)  # Listeyi DataFrame'e dntrme
x = veriler.iloc[:, 1:4]  # imdi "iloc" yntemini kullanabilirsiniz
x
# bamsz deikenler
```

```python
y=veriler.iloc[:,4:]
y
# baml deiken
```

```python
# verileri eitim ve test veri seti olarak ayrma
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
# veriliern leklendirilmesi:
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
# eitim setinide fit ile : modeli ren transform ile : renileni uygula
X_train=sc.fit_transform(x_train)
X_test=sc.transform(x_test) # test veri seti stnde eitim yapma modeli uygula
```

```python
X_train
```

```python
y_train=y_train.values
y_train
```

LOGISTIC REGRESYON

```python
from sklearn.linear_model import LogisticRegression

logr=LogisticRegression(random_state=0)
logr.fit(X_train,y_train)
y_pred=logr.predict(X_test)
y_pred
```

```python
y_test
```

image.png

```python
from sklearn.metrics import confusion_matrix
conf_m= confusion_matrix(y_test,y_pred)
conf_m # doru says :0 yanl 8
```

KNN


```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,metric="minkowski")

knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
y_pred
```
