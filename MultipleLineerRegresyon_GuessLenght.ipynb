```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Gender verisini çekin
gender = data.iloc[:, 4].values

# LabelEncoder kullanarak cinsiyet verisini sayısal hale getirin
le = LabelEncoder()
gender_encoded = le.fit_transform(gender)
print("LabelEncoded Cinsiyet Verisi:")
print(gender_encoded)

# OneHotEncoder kullanarak sayısal cinsiyet verisini one-hot encoding yapın
ohe = OneHotEncoder(sparse=False)  # sparse=False ile seyrek matrisi önler
gender_onehot = ohe.fit_transform(gender_encoded.reshape(-1, 1))
print("One-Hot Encoded Cinsiyet Verisi:")
print(gender_onehot)
```

```python
sol=data.iloc[:,:3]
sol
```

```python
# Convert the numpy arrays to DataFrames
sol_df = pd.DataFrame(sol)
sag_df = pd.DataFrame(sag)

# Concatenate the DataFrames
veri = pd.concat([sol_df, sag_df], axis=1)
veri
```
