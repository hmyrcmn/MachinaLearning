```python
# "cap-shape", "cap-color", "ring-number" ve "ring-type" sÃ¼tunlarÄ±nÄ± Ã¶zellikler olarak seÃ§in
X = dataset.loc[:, ["cap-shape", "cap-color", "ring-number", "ring-type"]]

# "class" sÃ¼tununu etiket olarak seÃ§in
y = dataset['class']
```

```python
from sklearn.preprocessing import LabelEncoder

# LabelEncoder nesnesi oluÅŸturun
label_encoder = LabelEncoder()

# Ã–zelliklerin (X) her bir sÃ¼tununu dÃ¶nerek dÃ¶nÃ¼ÅŸÃ¼m iÅŸlemini gerÃ§ekleÅŸtirin
for column in X.columns:
    X[column] = label_encoder.fit_transform(X[column])

# Etiketleri (y) doÄŸrudan LabelEncoder kullanarak dÃ¶nÃ¼ÅŸtÃ¼rÃ¼n
y = label_encoder.fit_transform(y)
```

```python
# Print X (features)
print("X (Features):")
print(X)

# Print y (labels)
print("\n y (Labels):")
print(y)
```

```python
from sklearn.model_selection import train_test_split

# Veri setini eÄŸitim ve test kÃ¼melerine bÃ¶lmek iÃ§in train_test_split kullanÄ±n
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# test_size, veri setinin hangi oranÄ±nÄ±n test seti olarak ayrÄ±lacaÄŸÄ±nÄ± belirler.
# random_state, veri setini rastgele bÃ¶lerken kullanÄ±lan bir tohum deÄŸerdir (seed value).

# EÄŸitim ve test kÃ¼melerini oluÅŸturduktan sonra, bu kÃ¼meleri makine Ã¶ÄŸrenimi modelinizi eÄŸitmek ve test etmek iÃ§in kullanabilirsiniz.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import RidgeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

# Logistic Regression modeli oluÅŸturma
logistic_model = LogisticRegression()

# Ridge Classifier modeli oluÅŸturma
ridge_model = RidgeClassifier()

# Decision Tree Classifier modeli oluÅŸturma
decision_tree_model = DecisionTreeClassifier()

# Gaussian Naive Bayes modeli oluÅŸturma
naive_bayes_model = GaussianNB()

# MLP (Multi-Layer Perceptron) Classifier modeli oluÅŸturma
mlp_model = MLPClassifier()

# Bu kod, her bir model iÃ§in bir model nesnesi oluÅŸturur ve bu nesneleri kullanarak eÄŸitim iÅŸlemine geÃ§ebilirsiniz.
```

```python
# Logistic Regression modelini eÄŸitme
logistic_model.fit(X_train, y_train)

# Ridge Classifier modelini eÄŸitme
ridge_model.fit(X_train, y_train)

# Decision Tree Classifier modelini eÄŸitme
decision_tree_model.fit(X_train, y_train)

# Gaussian Naive Bayes modelini eÄŸitme
naive_bayes_model.fit(X_train, y_train)

# MLP Classifier modelini eÄŸitme
mlp_model.fit(X_train, y_train)
```

```python
# Logistic Regression modeli ile tahmin yapma
logistic_predictions = logistic_model.predict(X_test)

# Ridge Classifier modeli ile tahmin yapma
ridge_predictions = ridge_model.predict(X_test)

# Decision Tree Classifier modeli ile tahmin yapma
decision_tree_predictions = decision_tree_model.predict(X_test)

# Gaussian Naive Bayes modeli ile tahmin yapma
naive_bayes_predictions = naive_bayes_model.predict(X_test)

# MLP Classifier modeli ile tahmin yapma
mlp_predictions = mlp_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

# Logistic Regression modeli iÃ§in sÄ±nÄ±flandÄ±rma raporu oluÅŸturma
logistic_report = classification_report(y_test, logistic_predictions)

# Ridge Classifier modeli iÃ§in sÄ±nÄ±flandÄ±rma raporu oluÅŸturma
ridge_report = classification_report(y_test, ridge_predictions)

# Decision Tree Classifier modeli iÃ§in sÄ±nÄ±flandÄ±rma raporu oluÅŸturma
decision_tree_report = classification_report(y_test, decision_tree_predictions)

# Gaussian Naive Bayes modeli iÃ§in sÄ±nÄ±flandÄ±rma raporu oluÅŸturma
naive_bayes_report = classification_report(y_test, naive_bayes_predictions)

# MLP Classifier modeli iÃ§in sÄ±nÄ±flandÄ±rma raporu oluÅŸturma
mlp_report = classification_report(y_test, mlp_predictions)

# SÄ±nÄ±flandÄ±rma raporlarÄ±nÄ± yazdÄ±rma
print("Logistic Regression Classification Report:")
print(logistic_report)

print("\nRidge Classifier Classification Report:")
print(ridge_report)

print("\nDecision Tree Classifier Classification Report:")
print(decision_tree_report)

print("\nGaussian Naive Bayes Classification Report:")
print(naive_bayes_report)

print("\nMLP Classifier Classification Report:")
print(mlp_report)
```
